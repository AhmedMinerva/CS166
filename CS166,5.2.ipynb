**Imagine a graph representing hockey players and hockey clubs. Each node is either a player or a club and each edge indicates that a player has played for that club at some point in the past. Explain why this graph is a bipartite graph**

A bipartiite graph is a graph whose nodes can be divided into two (or n) groups so that no edge connects nodes within each group. In this case, the hokey players will be a group and the hokey clubs will be another group. There will be no links between the hokey clubs with other clubs as it’s impossible for a club to represent the relation of playing at another club. This is true for players nodes as well.

```python
#Sayama Exercise 15.3.

f, axs = plt.subplots(2,2,figsize=(5,5))

subplot(2, 2, 1)
h=nx.complete_graph(4)
nx.draw_random(h)
title('4_complete')


subplot(2, 2, 2)
c=nx.complete_graph(5)
nx.draw_random(c)
title('5_complete')


subplot(2, 2, 3)
B = nx.Graph()
# Add nodes with the node attribute "bipartite"
B.add_nodes_from([1, 2], bipartite=0)
B.add_nodes_from(['a', 'b', 'c'], bipartite=1)
# Add edges only between nodes of opposite node sets
B.add_edges_from([(1, 'a'), (1, 'b'), (1, 'c'), (2, 'a'), (2, 'b'),(2, 'c')])

nx.draw_random(B)
title('bipartite')



subplot(2, 2, 4)
B = nx.Graph()
# Add nodes with the node attribute "bipartite"
B.add_nodes_from([1, 2, 3], bipartite=0)
B.add_nodes_from(['a', 'b', 'c'], bipartite=1)
# Add edges only between nodes of opposite node sets
B.add_edges_from([(1, 'a'), (1, 'b'), (1, 'c'), (2, 'a'), (2, 'b'),(2, 'c'), (3, 'a'), (3, 'b'),(3, 'c')])
#here a graph whose nodes can be divided into two (or n) groups so that no edge connects nodes within each group.

nx.draw_random(B)
title('bipartite')

show()
```
